```
url = "https://geodata.ucdavis.edu/gadm/gadm4.1/json/gadm41_IDN_2.json.zip"
```

```
gdf = gpd.read_file(url)
gdf.head()
```

```
gdf.columns
```

```
sorted(gdf["NAME_1"].unique())
```

```
prov_sulampua = [
    "SulawesiBarat",
    "SulawesiSelatan",
    "SulawesiTengah",
    "SulawesiTenggara",
    "SulawesiUtara",
    "Gorontalo",
    "Maluku",
    "MalukuUtara",
    "Papua",
    "PapuaBarat"
]

sulampua = gdf[gdf["NAME_1"].isin(prov_sulampua)].copy()
```

```
print(sulampua["NAME_1"].unique())
print(len(sulampua))
```
